```python
X=df[['R&D Spend','Administration','Marketing Spend']]   #split the data columns
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
y=df['Profit']
```

```python
y.shape
```

```python
## (i) Constructing different regression algorithms
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
## (ii) Split dataset into training set and test set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
 ## (iii) Construct different Regression algorithms

# Linear Regression
lr_model=LinearRegression()
lr_model.fit(X_train,y_train)
```

```python
# Decision Tree Regression
dt_model=DecisionTreeRegressor()
dt_model.fit(X_train,y_train)
```

```python
# Random Forest Regression
rf_model=RandomForestRegressor()
rf_model.fit(X_train,y_train)
```

```python
## (iv) Calculating different regression metrics

# Function to calculate and print regression metrics
def calculate_metrics(model,X,y,set_name):
    y_pred=model.predict(X)
    mse=mean_squared_error(y,y_pred)
    r2=r2_score(y,y_pred)
    print(f'{set_name}Mean Squared Error: {mse}')
    print(f'{set_name}R-squared: {r2}')
    print('\n')

# Calculating different regression metrics
calculate_metrics(lr_model,X_train,y_train,"Train Set (Linear Regression)")
calculate_metrics(lr_model,X_test,y_test,"Test Set (Linear Regression)")

calculate_metrics(dt_model,X_train,y_train,"Train Set (Decision Tree)")
calculate_metrics(dt_model,X_test,y_test,"Test Set (Decision Tree)")

calculate_metrics(rf_model,X_train,y_train,"Train Set (Random Forest)")
calculate_metrics(rf_model,X_test,y_test,"Test Set (Random Forest)")



```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
mean_absolute_error(y_test,y_pred)
mean_absolute_percentage_error(y_test,y_pred)
lr.score(X_test,y_test)
```

```python
dt=DecisionTreeRegressor()
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
y_pred
```

```python
mean_absolute_error(y_test,y_pred)
mean_absolute_percentage_error(y_test,y_pred)
dt.score(X_test,y_test)
```

```python
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
y_pred
```

```python
mean_absolute_error(y_test,y_pred)
mean_absolute_percentage_error(y_test,y_pred)
rf.score(X_test,y_test)
```
